Example of focus stacking with:

```python
from shinestacker.config.constants import constants
from shinestacker.algorithms import *
```

```python
job = StackJob("job", ".", input_path="input/amber-flies")
job.add_action(CombinedActions("amber-flies-align", [AlignFrames()]))
job.add_action(FocusStack("amber-flies-stack", PyramidStack()))
job.run()
```

```python
from shinestacker.algorithms.utils import img_8bit, read_img
import matplotlib.pyplot as plt
import cv2

img = cv2.cvtColor(img_8bit(read_img("amber-flies-stack/stack_flies_0000.jpg")), cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 5))
plt.imshow(img, 'gray')
plt.show()
```
